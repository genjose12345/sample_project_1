## matplotlib example workflow

```python
import os
import matplotlib.pyplot as plt
%matplotlib inline

# Create images directory if it doesn't exist
if not os.path.exists('images'):
    os.makedirs('images')

# 1. prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. setup plot
fig, ax = plt.subplots(figsize=(10, 10))#with,height

# 3. plot data
ax.plot(x, y)


# 4. customize plot
ax.set(title="simple plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. save and show
fig.savefig("images/sample-plot.png")
plt.show()  # Added to display the plot in Jupyter notebook
```

```python
# First make sure matplotlib inline is set at the top of your notebook
%matplotlib inline
import matplotlib.pyplot as plt

# Create dictionary with correct spelling
nut_butter_prices = {
    "almond butter": 10,
    "peanut butter": 8,    # fixed spelling of "peanut"
    "cashew butter": 12    # made naming consistent
}

# Create the plot
fig, ax = plt.subplots(figsize=(8, 6))  # added figure size for better visibility
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())

# Customize the plot
ax.set(title="Dan's Nut Butter Store", 
       ylabel="Price ($)")

# Show the plot
plt.show()
```

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
#pyplot method
plt.close()
over_50.plot(kind = 'scatter',
            x='age',
             y='chol',
             c='target')
plt.show()  
```

```python
#00 method
plt.close()
fix,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind ='scatter',
                    x='age',
                    y='chol',
                    c='target',
                    ax=ax)
ax.set_xlim([45,100])
plt.show()
```

```python
plt.close()
##00 method from scratch 
fig,ax = plt.subplots(figsize=(10,6))

#plot the data
scatter = ax.scatter(x=over_50["age"],
                    y= over_50["chol"],
                    c=over_50["target"]);

#Customize the plot
ax.set(title="Heart disease and cholesterol levels",
      xlabel="Age",
      ylabel="Cholesterol");

#add a lenged
ax.legend(*scatter.legend_elements(),title="Target");

#add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
plt.show()
```

```python
over_50.target.values
```

```python
over_50.head()
```

```python
#subplot of chol age thalach
plt.close()
fig,(ax0,ax1) = plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,10),
                            sharex = True)
#add data to ax0
scatter =  ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])
#customize ax0
ax0.set(title = "Heart dsease and cholesterol leels",
       ylabel="Cholesterol");

#add a legned to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

#add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle= '--');
#add data to ax1
scatter = ax1.scatter(x= over_50["age"],
                     y= over_50["thalach"],
                     c= over_50["target"])
#customize ax1
ax1.set(title="Heart Disease and max heart rate",
       xlabel = "Age",
       ylabel="Max Heart Rate");

#add a lenged to ax1
ax1.legend(*scatter.legend_elements(),title="Target")

#add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
          linestyle="--")
#add a figure
fig.suptitle("Hear Disease Analysis", fontsize =16, fontweight="bold")
plt.show()
```

```python
plt.close()

#set style
plt.style.use('seaborn-v0_8-whitegrid')

##00 method from scratch 
fig,ax = plt.subplots(figsize=(10,6))

#plot the data
scatter = ax.scatter(x=over_50["age"],
                    y= over_50["chol"],
                    c=over_50["target"],
                    cmap = "winter");

#Customize the plot
ax.set(title="Heart disease and cholesterol levels",
      xlabel="Age",
      ylabel="Cholesterol");

#add a lenged
ax.legend(*scatter.legend_elements(),title="Target");

#add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
plt.show()
```

```python
# customzing the y and x axist

#subplot of chol age thalach
plt.close()
fig,(ax0,ax1) = plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,10),
                            sharex = True)
#add data to ax0
scatter =  ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                      cmap = "winter")
#customize ax0
ax0.set(title = "Heart dsease and cholesterol leels",
       ylabel="Cholesterol");
#change the x axis
ax0.set_xlim([50,80])

#add a legned to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

#add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle= '--');
#add data to ax1
scatter = ax1.scatter(x= over_50["age"],
                     y= over_50["thalach"],
                     c= over_50["target"],
                     cmap= "winter")
#customize ax1
ax1.set(title="Heart Disease and max heart rate",
       xlabel = "Age",
       ylabel="Max Heart Rate");

#change x axist limits
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

#add a lenged to ax1
ax1.legend(*scatter.legend_elements(),title="Target")

#add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
          linestyle="--")
#add a figure
fig.suptitle("Hear Disease Analysis", fontsize =16, fontweight="bold")
plt.show()
```
