## 0. an end to end scikit-learn workflow

```python
# 1 get the data ready
import pandas as pd
import numpy as np
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
#create x (feautres matrix)
x = heart_disease.drop("target", axis = 1)

#create y Label
y = heart_disease["target"]
```

```python
# 2 choose the right model and hyper parameter
from sklearn.ensemble import RandomForestClassifier
clf= RandomForestClassifier()

#we'll keep the default hyperparameters
clf.get_params()
```

```python
#3 fit the model to the traning data 
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
clf.fit(x_train,y_train)
```

```python
#y_label = clf.predict(np.array([0,2,3,4]))
```

```python
y_pred = clf.predict(x_test)
y_pred
```

```python
y_test
```

```python
#4. evaluate the model on the training data and test data
clf.score(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
#5 imporve a model
#try different amout of n_estators
np.random.seed(45)
for i in range(10,100,10):
    print(f"trying model with {i} estimators...")
    clf = RandomForestClassifier(n_estimators=i).fit(x_train,y_train)
    print(f"Model accuracy on test set: {clf.score(x_test,y_test)* 100:.2f}%")
    print("")
```

```python
#6 save a model and load it 
import pickle
pickle.dump(clf,open("random_forest_model_1.pkl","wb"))
```

```python
loaded_model = pickle.load(open("random_forest_model_1.pkl","rb"))
loaded_model.score(x_test,y_test)
```

### what if there were missing value?
1. fill them with some value(also known as imputation)
2. remove sampels with missing data altogether



```python
#import car sales missing data
car_sales_missing = pd.read_csv("car-sales-missing-data.csv")
car_sales_missing.head()
```

### option 1 : fill missing data with pandsa

```python
#fill the make column
car_sales_missing["Make"].fillna("missing", inplace = True)

#filling the colour column
car_sales_missing["Colour"].fillna("missing",inplace=True)

#fill the Odometer(KM) column
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean(),inplace=True)

#fill the doors column
car_sales_missing["Doors"].fillna(4,inplace=True)
```
