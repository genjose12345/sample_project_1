## matplotlib example workflow

```python
import os
import matplotlib.pyplot as plt
%matplotlib inline

# Create images directory if it doesn't exist
if not os.path.exists('images'):
    os.makedirs('images')

# 1. prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. setup plot
fig, ax = plt.subplots(figsize=(10, 10))#with,height

# 3. plot data
ax.plot(x, y)


# 4. customize plot
ax.set(title="simple plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. save and show
fig.savefig("images/sample-plot.png")
plt.show()  # Added to display the plot in Jupyter notebook
```

```python
# First make sure matplotlib inline is set at the top of your notebook
%matplotlib inline
import matplotlib.pyplot as plt

# Create dictionary with correct spelling
nut_butter_prices = {
    "almond butter": 10,
    "peanut butter": 8,    # fixed spelling of "peanut"
    "cashew butter": 12    # made naming consistent
}

# Create the plot
fig, ax = plt.subplots(figsize=(8, 6))  # added figure size for better visibility
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())

# Customize the plot
ax.set(title="Dan's Nut Butter Store", 
       ylabel="Price ($)")

# Show the plot
plt.show()
```

```python
nut_butter_prices = {"almond butter" : 10,
                     "penut butter" : 8,
                     "cashew" : 12}
fig,ax = plt .subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title = "dan's nut butter store",ylabel="Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
plt.show()
```

```python
# subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))  # figsize needs two values: (width, height)
#plot to eahc different fig
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10),np.random.random(10)),
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
```

```python
# subplots ption 2
fig,ax = plt.subplots(nrows=2,
                     ncols=2,
                     figsize=(10,5))
#plot to each differeent index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

```python
#00 method
plt.close()
fix,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind ='scatter',
                    x='age',
                    y='chol',
                    c='target',
                    ax=ax)
plt.show()
```
